```python
# Check if all Dividends and Stock Splits are zero
are_dividends_zero = (data['Dividends'] == 0).all()
are_splits_zero = (data['Stock Splits'] == 0).all()

if are_dividends_zero:
    print("All Dividend values are zero.")
else:
    print("There are non-zero Dividend values.")

if are_splits_zero:
    print("All Stock Split values are zero.")
else:
    print("There are non-zero Stock Split values.")
```

```python
non_zero_dividends = data[data['Dividends'] != 0]
non_zero_splits = data[data['Stock Splits'] != 0]

print("Non-zero Dividend rows:")
print(non_zero_dividends)

print("Non-zero Stock Split rows:")
print(non_zero_splits)
```

```python
data['Adjusted Close'] = data['Close'] / data['Stock Splits'].replace(0, 1).cumprod()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(data['Date'], data['Close'], label='Close Price')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('AAPL Stock Price Over Time')
plt.legend()
plt.show()
```

```python
total_dividends = data['Dividends'].sum()
print(f"Total Dividends Paid: {total_dividends}")
```

```python
total_splits = data['Stock Splits'].sum()
print(f"Total Stock Splits: {total_splits}")
```

```python
data.to_csv('Processed_Stock_Data.csv', index=False)
print("Processed data saved to 'Processed_Stock_Data.csv'")
```
